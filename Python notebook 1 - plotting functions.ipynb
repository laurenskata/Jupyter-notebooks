```python
import matplotlib.pyplot as plt
plt.plot([1,5,3,4,5])
plt.ylabel('Y-axis')
plt.xlabel('X-axis')
plt.title('Matlab copy')
plt.show()
```

```python
import numpy as np

def f(t):
    return 2*np.sin(2*np.pi*t)+0.6*np.cos(2*np.pi*t)
t = np.arange(0.0,10.0,0.001)

plt.plot(t,f(t))
plt.xlabel('time(s)')
plt.ylabel('Amplitude')
plt.show()
```
