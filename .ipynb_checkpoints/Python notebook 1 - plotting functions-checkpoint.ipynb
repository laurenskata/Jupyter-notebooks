```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return 2*np.sin(2*np.pi*t)+0.6*np.cos(2*np.pi*t)
t = np.arange(0.0,10.0,0.001)

plt.plot(t,f(t))
plt.xlabel('time(s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return 2*np.sin(2*np.pi*t)+0.6*np.cos(2*np.pi*t)
def g(t):
    return np.cos(2*np.pi*0.5*t)
t = np.arange(0.0,10.0,0.001)

#overlaying multiple graphs in 1
plt.plot(t,f(t))
plt.plot(t,g(t))
plt.xlabel('time(s)')
plt.ylabel('Amplitude')
plt.title('2 functions in  one plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return 2*np.sin(2*np.pi*t)+0.6*np.cos(2*np.pi*t)
def g(t):
    return np.cos(2*np.pi*0.5*t)
t = np.arange(0.0,10.0,0.001)

#creating the figure
plt.figure(1)

#subplot 1
plt.subplot(211)
plt.plot(t,f(t))
plt.title('2 functions in  one plot')
plt.ylabel('Amplitude')

#subplot 2
plt.subplot(212)
plt.plot(t,g(t))
plt.xlabel('time(s)')
plt.ylabel('Amplitude')
plt.show()
```
